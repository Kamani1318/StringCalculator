```python
# Generate a dataset of 20+ random numbers. Apply suitable dataset to draw the following plots:
x = np.random.randint(10, size=(20))
y1 = np.random.randint(10,size= (20))
y2 = np.random.randint(10,size= (20))
```

```python
# Line Plots
plt.plot(x)
plt.show()
```

```python
# Histograms
plt.hist(x)
plt.show()
```

```python
# Bar plots
plt.bar(x,y1)
```

```python
# Side by side bar charts
plt.bar(x -0.2, y1, width=0.4, color = 'purple')
plt.bar(x +0.2, y2, width=0.4, color = 'green')
```

```python
# Stacked bar charts
plt.bar(x, y1, color='red')
plt.bar(x, y2, bottom=y1, color='blue')
plt.show()
```

```python
# sine curvex 
import math
x = np.arange(0,2*math.pi,0.01)
y = [math.sin(i) for i in x]
plt.plot(x,y)
plt.show()
```

# 3D scatter plots 

```python
matrix = np.hstack((x.reshape(-1, 1), y1.reshape(-1, 1),y2.reshape(-1, 1)))
data = pd.DataFrame(matrix, columns = ['x','y1','y2'])
```

```python
import seaborn as sns
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y1, y2)
ax.set_xlabel('X Label')
ax.set_ylabel('Y1 Label')
ax.set_zlabel('Y2 Label')
ax.set_title('3D Scatter Plot')
```

```python
sns.violinplot(x="x", y="y1", data=data)
plt.title("Violin plot example")
plt.xlabel("x1")
plt.ylabel("y1")
```
