```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
marks = np.random.randint(100,size = (20,5))
grade = np.random.randint(10,size = (20))
students = pd.DataFrame(marks, columns = ['english','math','hindi','physics','biology'])
students['grade'] = grade
students.head()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(20, 6))
sns.scatterplot(data=students, x='english',y = 'grade', ax=axs[0])
sns.scatterplot(data = students, x = 'math', y = 'grade', ax = axs[1])
sns.scatterplot(data = students, x = 'hindi', y = 'grade', ax = axs[2])
sns.scatterplot(data = students, x = 'physics', y = 'grade', ax = axs[3])
sns.scatterplot(data = students, x = 'biology', y = 'grade', ax = axs[4])
```

```python

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(marks,grade)
```

```python
grade_pred = reg.predict(marks)
print('The MLR coefficients are : ', reg.coef_)
print('The MLR intercept is : ', reg.intercept_)
students['grade_pred'] = grade_pred
inputs = []
for i in range(5):
    num = int(input(f"Enter input {i+1}: "))
    inputs.append(num)
print("The output for the new data point is:",reg.predict([[input]]))
```

```python
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(20, 6))
sns.lineplot(data=students, x='english',y = 'grade_pred', ax=axs[0])
sns.lineplot(data = students, x = 'math', y = 'grade_pred', ax = axs[1])
sns.lineplot(data = students, x = 'hindi', y = 'grade_pred', ax = axs[2])
sns.lineplot(data = students, x = 'physics', y = 'grade_pred', ax = axs[3])
sns.lineplot(data = students, x = 'biology', y = 'grade_pred', ax = axs[4])
```
