```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
x = np.random.randint(100,size = (10))
y = np.random.randint(100,size = (10))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x, y)
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(x.reshape(-1,1),y.reshape(-1,1))
```

```python
#  Simple Linear Regression Coefficients
m = reg.coef_
b = reg.intercept_
m = m.item()
m
```

```python
y_pred = m*x + b
```

```python
new_point = int(input('Enter the value of a new data point: '))
print("The output for the new data point is:",reg.predict([[new_point]]))
```

```python
%matplotlib inline
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x,y,color = 'red')
plt.plot(x,y_pred,color = 'blue')
```

```python
def sum_square_error(expected,actual):
    sse = 0
    for i in range(0,len(actual)):
        sse = sse + (expected[i] - actual[i])**2
    return sse
```

```python
SSE = sum_square_error(y_pred,y)
SSE
```

```python
def sum_square_regression(expected,actual):
    ssr = 0
    actual_mean = actual.mean()
    for i in range(0,len(actual)):
        ssr = ssr + (expected[i]-actual_mean)**2
    return ssr
```

```python
SSR = sum_square_regression(y_pred,y)
SSR
```

```python
def sum_square_total(expected,actual):
    sst = 0
    actual_mean = actual.mean()
    for i in range(0,len(actual)):
        sst = sst + (actual[i]-actual_mean)**2
    return sst
```

```python
SST = sum_square_total(y_pred,y)
SST
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train, y_test = train_test_split(x,y ,random_state=42, test_size=0.2, shuffle=True)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train, y_train)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression  
regressor= LinearRegression() 
regressor.fit(x_train.reshape(-1,1), y_train.reshape(-1,1))  
m1 = regressor.coef_
b1 = regressor.intercept_
regressor.predict([[33]])
y_pred_train = (m1*x_train + b1).flatten()
```

```python
%matplotlib inline
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x_train,y_train,color = 'red')
plt.plot(x_train,y_pred_train,color = 'blue')
```

```python
y_pred_train = y_pred_train.flatten()
```

```python
print(y_pred_train)
print(y_train)
```

```python
SSE_train = sum_square_error(y_pred_train,y_train)
SSR_train = sum_square_regression(y_pred_train,y_train)
SST_train = sum_square_total(y_pred_train,y_train)
print("SSE for training data: ",SSE_train)
print("SSR for training data: ",SSR_train)
print("SST for training data: ",SST_train)
```
