```python
# To take in a random number as input and check if the number is even or odd in a function
def check_number(num):
    if num%2 == 0:
        print('Number is even')
    else:
        print('Number is odd')
```

```python
check_number(int(input('Enter a random number: ')))
```

```python
def enter_inputs():
    list = []
    n = int(input("Enter the number of items: "))
    for i in range(n):
        x = int(input("Enter number: "))
        list.append(x)
    return list
    
def check_num(list):
    for i in list:
        if i%2 == 0:
            print("Number ",i," is even")
        else:
            print("Number ",i," is odd")
            

list = enter_inputs()
check_num(list)
```

```python
cereals = pd.read_csv('cereal.csv')
cereals.head()
```

```python
# Tabulate the following attributes: mfr and type of cereals
pd.crosstab(cereals.mfr,cereals.type)
```

```python
# Display the 5 number summary for all nutritional attributes starting from calories
cereals.describe().drop(columns = ['shelf','weight','cups','rating'])
# Note: a value of -1 indicates there are missing values in that particluar column
```

```python
 # For missing values find and replace with arithmetic mean of the attributes.
cereals[cereals == -1].sum()
```

```python
carbo_mean = cereals['carbo'].mean()
sugars_mean = cereals['sugars'].mean()
potass_mean = cereals['potass'].mean()
cereals['carbo'].replace(-1,carbo_mean,inplace = True)
cereals['sugars'].replace(-1,carbo_mean,inplace = True)
cereals['potass'].replace(-1,carbo_mean,inplace = True)
```

```python
#  Find and replace outliers with median.
def find_outliers_IQR(df):
    q1=df.quantile(0.25)

    q3=df.quantile(0.75)
    IQR = (q3-q1)
    return IQR,q3,q1

IQR,q3,q1 = find_outliers_IQR(cereals)
new_cereals = cereals.drop(columns = ['name','mfr','type'])
new_cereals
non_outliers = new_cereals[(new_cereals>(q1 - 1.5*IQR)) | (new_cereals<(q3 + 1.5*IQR))]
non_outliers.fillna(non_outliers.median())
```
